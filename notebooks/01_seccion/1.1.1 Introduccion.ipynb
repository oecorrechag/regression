## 1.1.1.1 ¿Qué es el análisis de regresión?

Es un técnica estadística para modelar la relación entre variables. Su objetivo principal es entender cómo las variables independientes afectan a la variable dependiente y predecir o estimar el valor de la variable dependiente en función de las variables independientes.

- Y > Variable respuesta (dependiente)
- X > Variables explicativas, independiente regresores o covariables

## 1.1.1.2 ¿En dónde se aplica en análisis de regresión?

- Ingeniería
- Física
- Química
- Economía
- Administración
- Ciencias biológicas
- Ciencias de la salud
- Ciencias sociales
- Mercadeo

## 1.1.1.3 Ejemplo

Un profesional está estudiando la relación que existe entre el tiempo de servicio de una máquina dispensadora de refrescos y el número de refrescos suministrados

## 1.1.1.4 Modelo base

$$ y = \beta_{0} + \beta_{1}x + e $$

con:

- E(e) = 0
- Var(e) = $\sigma^2$

## 1.1.1.6 Tipos de estudios

- Estudios Descriptivos:

    - Estudios de Casos: Describen detalladamente un caso o un grupo de casos.
    - Estudios Transversales: Recolectan datos de una población o muestra en un punto específico en el tiempo.
    - Estudios Longitudinales: Siguen a una población o muestra a lo largo del tiempo para analizar cambios y tendencias.

- Estudios Analíticos:

    - Estudios de Cohorte: Observan a un grupo de individuos a lo largo del tiempo, comparando aquellos expuestos y no expuestos a un factor específico.
    - Estudios de Casos y Controles: Comparan individuos con una enfermedad o condición específica (casos) con aquellos sin la condición (controles) para identificar posibles factores de riesgo.
    - Ensayos Clínicos: Pruebas controladas y aleatorias de intervenciones médicas o tratamientos.

- Estudios Experimentales:

    - Experimentos Aleatorios: Asignan aleatoriamente participantes a diferentes grupos para investigar el efecto de una variable independiente sobre una variable dependiente.
    - Experimentos Cuasiexperimentales: Similar a los experimentos aleatorios, pero con asignación no aleatoria de participantes.

- Estudios Observacionales:

    - Estudios Observacionales Prospectivos: Observan a individuos en el presente y los siguen en el futuro.
    - Estudios Observacionales Retrospectivos: Utilizan datos históricos para analizar eventos pasados.

- Estudios Exploratorios:

    - Estudios Piloto: Investigaciones preliminares pequeñas diseñadas para probar la viabilidad y factibilidad de un estudio más grande.
    - Estudios Exploratorios: Buscan comprender fenómenos poco conocidos y generar hipótesis para investigaciones futuras.

- Metaanálisis:

    - Se trata de un estudio estadístico que combina y analiza los resultados de múltiples estudios independientes sobre un tema específico, proporcionando una síntesis cuantitativa de la evidencia disponible.
